```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import sys
import os
```

# Imported open cv module- **OpenCV**  used for a  real-time computer vision.

```python
original_avengers_image = cv2.imread('../input/avengers/avengers.jpg')
```

> AS i am using kaggle notebook on browser so i cant use opencv function imshow as it will open new window but in browser it wont supported 

```python
plt.imshow(original_avengers_image)
plt.title("MATPLOTLIB IMAGE")
plt.show()
```

needs to chenge this color from BGR TO RGB

```python
plt.imshow(cv2.cvtColor(original_avengers_image,cv2.COLOR_BGR2RGB))
plt.title("ORIGINAL IMAGE")
plt.show()
```

**original image converts to gray scale function** by eliminating the hue and saturation information while retaining the luminance

```python
grey_avengers_image=cv2.cvtColor(original_avengers_image,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(grey_avengers_image, cv2.COLOR_BGR2RGB))
plt.title("GRAYSCALE IMAGE")
plt.show()
```

**bitwise_not()** -This function will help the image to keep the properties as same by providing the masking to it.

it convert pixel value to zero which are greater than 0 and zero value convert into 255

rgb (255,255,255) white to black 
and black to white

```python
invert_avengers_image=cv2.bitwise_not(grey_avengers_image)
plt.imshow(cv2.cvtColor(invert_avengers_image, cv2.COLOR_BGR2RGB))
plt.title("LAYERED NEGATIVE  IMAGE")
plt.show()
```

**GaussianBlur()-** This function is used to modify the image by sharpening the edges of the image, smoothen the image, and will minimize the blurring property.

```python
blur_avengers_image=cv2.GaussianBlur(invert_avengers_image, (21,21),sigmaX=0, sigmaY=0)

```

 **divide()**- This function is used for the normalization of the image as it doesn’t lose its previous properties.

```python
sketch=cv2.divide(grey_avengers_image,255-blur_avengers_image, scale=255.0)
plt.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
plt.title("PENCIL SKETCH")
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(original_avengers_image)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(grey_avengers_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(invert_avengers_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(blur_avengers_image,cmap="gray")
plt.axis("off")
plt.title("Blur Image")
plt.subplot(1,5,5)
plt.imshow(sketch,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
