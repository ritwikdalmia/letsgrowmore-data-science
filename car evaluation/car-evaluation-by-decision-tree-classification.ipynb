```python
x=dataset_read.drop(['class'], axis=1)
y=dataset_read['class']
```

**spiliting data of x and y for training and testing**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 40)
```

```python
print ('\n training data capacity for feature  65% ', x_train.shape,'\n testing data capacity 35% for feature ',x_test.shape)
```

```python
print ('\n training data capacity  for targeted 65%  ', y_train.shape,'\n testing data capacity  for targeted 35% ',y_test.shape)
```

```python
import category_encoders as ce
```

```python
# encode variables with ordinal encoding

encoder = ce.OrdinalEncoder(dataset_read[['buying', 'maintaince', 'doors', 'person_capacity', 'luggage_boot', 'safety']])


x_train = encoder.fit_transform(x_train)

x_test = encoder.transform(x_test)
```

```python
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# instantiate the DecisionTreeClassifier model with criterion gini index

classification_gini = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=0)
# fit the model
classification_gini.fit(x_train, y_train)
```

# predicting the model

```python

y_pred_classification_gini = classification_gini.predict(x_test)
```

# Check accuracy score with criterion gini index

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index:{:2f}'. format(classification_gini.score(x_test, y_test)*100))
```

Here, y_test are the actual class  and values y_pred_classification_gini are the predicted class value in the test-set.

```python
#confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_classification_gini)

print('Confusion matrix\n\n', cm)
```

```python
sns.heatmap(cm,annot=True,cmap='BuPu_r')
plt.rcParams["figure.figsize"]=(10,3)
plt.xlabel('predicted value')
plt.ylabel('class')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_classification_gini))
```
